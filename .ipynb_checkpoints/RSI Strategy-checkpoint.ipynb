```python
import pandas as pd
import mplfinance as mpf
import numpy as np
from pandas_datareader import data as web
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yf.pdr_override()
```

```python
start = '2020-06-01'
end = '2021-06-01'
```

```python
data = web.get_data_yahoo('NVDA', start=start, end=end)
```

```python
data['change'] = data['Open'] - data['Close']
```

```python
data.head(10)
```

```python
data['gain'] = data['change'].mask(data['change'] < 0).fillna(0)
data['loss'] = abs(data['change'].mask(data['change'] > 0)).fillna(0)
```

```python
data.head(10)
```

```python
# data['avg gain'] = ((data['gain'].shift(1) * 13) + data['gain']) / 14
# data['avg loss'] = ((data['loss'].shift(1) * 13) + data['loss']) / 14
data['avg gain'] = data['gain'].rolling(14).mean()
data['avg loss'] = data['loss'].rolling(14).mean()
data['RS'] = data['avg gain'] / data['avg loss']
data['RSI'] = 100 - ( 100 / ( 1 + data['RS'] ) )
```

```python
data.head(20)
```

```python
data_plot = data[data.RSI.notnull()]
data_plot = data_plot[['RSI']]
data_plot
```

```python
plt.figure(figsize=(15, 10))
plt.title('RSI')
plt.plot(data_plot.index, data_plot[['RSI']])
plt.xlabel('Trading Dates')
plt.ylabel('Value')
plt.legend(('RSI'), loc='lower right')
plt.axhline(y = 70, color = 'r', linestyle = '-')
plt.axhline(y = 30, color = 'r', linestyle = '-')
plt.show()
```

```python
def create_rsi_position(data, overbought=70, oversold=30):
    data.loc[(data["RSI"].shift(1) > oversold) & (data["RSI"] <= oversold), 'action'] = 'buy'
    data.loc[(data["RSI"].shift(1) < overbought) & (data["RSI"] >= overbought), 'action'] = 'sell'

    data.loc[(data["RSI"].shift(1) <= oversold) & (data["RSI"] <= oversold), 'action'] = np.nan
    data.loc[(data["RSI"].shift(1) >= overbought) & (data["RSI"] >= overbought), 'action'] = np.nan

    data.loc[data['action'] == 'buy', 'marker_position'] = data['Low'] # * 0.98
    data.loc[data['action'] == 'sell', 'marker_position'] = data['High'] # * 1.02

# data
# data[data['action'] == 'buy']
# data[data['action'] == 'sell']

# data.loc[ (data["RSI"].shift(1) >= 70) & (data["RSI"] >= 70) & (data["action"] == 'sell') ]
# data.loc[ (data["RSI"].shift(1) <= 30) & (data["RSI"] <= 30) & (data["action"] == 'buy') ]
```

```python
data70_30 = data.copy()
create_rsi_position(data70_30)
```

```python
def rsi_plot(df, overbought=70, oversold=30):
    up_markers = []
    down_markers = []
    markers = []
    
    up_line = np.empty(df.shape[0])
    up_line.fill(overbought)
    down_line = np.empty(df.shape[0])
    down_line.fill(oversold)
    
    for index, row in df.iterrows():
        x = row['action']
        if x == 'buy':
            up_markers.append(row['marker_position'])
            down_markers.append(np.nan)
        elif x == 'sell':
            up_markers.append(np.nan)
            down_markers.append(row['marker_position'])
        else:
            up_markers.append(np.nan)
            down_markers.append(np.nan)
    
    up_plot = mpf.make_addplot(up_markers, type='scatter', marker='^', markersize=100, panel=0, color='green')
    down_plot = mpf.make_addplot(down_markers, type='scatter', marker='v', markersize=100, panel=0, color='red')
    
    rsi_plot = mpf.make_addplot(df['RSI'], panel=1, color='blue', title='RSI')
    overbought_plot = mpf.make_addplot(up_line, panel=1, color='orange')
    oversold_plot = mpf.make_addplot(down_line, panel=1, color='orange')
    
    plots = [up_plot, down_plot, rsi_plot, overbought_plot, oversold_plot]
    
    return plots
```

```python
a = rsi_plot(data70_30)
```

```python
mpf.plot(data, type='candle', style='yahoo', addplot=a, figsize=(10, 6))
```
