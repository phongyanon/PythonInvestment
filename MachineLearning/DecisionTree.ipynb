```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DecisionTreeExample.csv')
df
```

```python
# convert string data to number
d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)
df
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

plt.figure(figsize=(9,12))  # set plot size (denoted in inches)
tree.plot_tree(dtree, feature_names=features)
plt.show()
```

```python
print(dtree.predict([[40, 10, 7, 1]])) # [1] means Go, [0] manns No
```
