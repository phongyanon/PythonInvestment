```python
import pandas_datareader as web

start = '2020-01-01'
end = '2021-01-01'
```

```python
tsla = web.get_data_yahoo('TSLA', start=start, end=end)
tsla
```

```python
tsla_weekly = tsla.asfreq('W-Mon')
tsla_weekly
```

```python
tsla_monthly = tsla.resample('M').mean()
tsla_monthly
```
