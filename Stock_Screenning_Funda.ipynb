```python
import pandas as pd
import numpy as np
```

```python
path = 'https://drive.google.com/uc?export=download&id=1EqhMOcAPX7WfW1gObCPPsFq99D5RfA42'
```

```python
df = pd.read_excel(path)
```

```python
df.dtypes
```

```python
df.loc[df['P/E'] > 100].sort_values('P/E')
```

```python
df.loc[df['P/E'] > 100].sort_values(['P/E', 'P/BV'], ascending=[True, True])
```

```python
df['P/BV'] = df['P/BV'].astype(float)
```

```python
df[df['P/BV'] == '-']
```

```python
df = df.replace('-', np.nan)
```

```python
df[(df['P/E'] > 10) & (df['P/E'] < 100)].sort_values(['P/E','P/BV'],ascending = [True,True])
```
