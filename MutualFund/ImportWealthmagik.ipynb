```python
def getResponseSoup(url):
    res = requests.get(url)
    res.encoding = "utf-8"
    if res.status_code == 200:
        print("Successful")
    elif res.status_code == 404:
        print("Error 404 page not found")
    else:
        print("Not both 200 and 404")
        
    soup = BeautifulSoup(res.text, 'html.parser')
    return soup
```

```python
# data = soup.find("div", {"class":"namedesw7"})
# data = data.find_all("a", {"class": "namelink"})
"""
loop each funds to get link and request to link
data to collect:
- size
- historical profit each year
- management fee
- dividend yield
"""
fieldnames = ['name', 'size', 'frontend fee', 'backend fee', 'management fee', 'dividend yield', \
             'retrun 3 month', 'retrun 6 month', 'retrun 1 year', 'retrun 3 year', 'retrun 5 year', 'updated date']

rows = []
for name, symbol in equity_funds:
    profile_url = f"https://www.wealthmagik.com/FundInfo/FundProfile-{symbol}"
    fund_fee_url = f"https://www.wealthmagik.com/FundInfo/FundPurchaseFee-{symbol}"
    performance_url = f"https://www.wealthmagik.com/FundInfo/FundPerformance-{symbol}"
    
    profile_soup = getResponseSoup(profile_url)
    fund_fee_soup = getResponseSoup(fund_fee_url)
    performance_soup = getResponseSoup(performance_url)
    
    # get fund size
    fund_size = profile_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_lblTotalNetAsset"})
    fund_size = int(''.join(fund_size.string.split(',')))
    
    # dividend yield
    dividend_yield = profile_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_lblDividendPolicy"})
    dividend_yield = bool(dividend_yield.string == u'จ่ายเงินปันผล')
    
    # updated date
    updated_date = profile_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_lblInfoDate"})
    updated_date = updated_date.string
    
    # front-end fee
    frontend_fee = fund_fee_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_lblFrontFee"})
    frontend_fee = frontend_fee.string
    
    # back-end fee
    backend_fee = fund_fee_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_lblBackFee"})
    backend_fee = backend_fee.string
    
    # management fee
    management_fee = fund_fee_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_lblManagementFee"})
    management_fee = management_fee.string

    # Return 3 month
    return_3_month = performance_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_gridReturn_ctl02_lbl3Month"})
    if (not (return_3_month.string and return_3_month.string.strip())) or (return_3_month.string == '-'):
        return_3_month = 0
    else:
        return_3_month = float(return_3_month.string[:-1])
    
    # Return 6 month
    return_6_month = performance_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_gridReturn_ctl02_lbl6Month"})
    if (not (return_6_month.string and return_6_month.string.strip())) or (return_6_month.string == '-'):
        return_6_month = 0
    else:
        return_6_month = float(return_6_month.string[:-1])
    
    # Return 1 year
    return_1_year = performance_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_gridReturn_ctl02_lbl1Year"})
    if (not (return_1_year.string and return_1_year.string.strip())) or (return_1_year.string == '-'):
        return_1_year = 0
    else:
        return_1_year = float(return_1_year.string[:-1])
    
    # Return 3 year
    return_3_year = performance_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_gridReturn_ctl02_lbl3Year"})
    if (not (return_3_year.string and return_3_year.string.strip())) or (return_3_year.string == '-'):
        return_3_year = 0
    else:
        return_3_year = float(return_3_year.string[:-1])
    
    # Return 5 year
    return_5_year = performance_soup.find("span", {"id": "ctl00_ContentPlaceHolder1_gridReturn_ctl02_lbl5Year"})
    if (not (return_5_year.string and return_5_year.string.strip())) or (return_5_year.string == '-'):
        return_5_year = 0
    else:
        return_5_year = float(return_5_year.string[:-1])
        
    row = {
        'name': name, 
        'size': fund_size,
        'frontend fee': frontend_fee,
        'backend fee': backend_fee, 
        'management fee': management_fee, 
        'dividend yield': dividend_yield,
        'retrun 3 month': return_3_month, 
        'retrun 6 month': return_6_month, 
        'retrun 1 year': return_1_year, 
        'retrun 3 year': return_3_year, 
        'retrun 5 year': return_5_year, 
        'updated date': updated_date
    }
    rows.append(row)
    
now_date = datetime.today().strftime('%Y_%m_%d')
with open(f'equity_fund_{now_date}.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)
    print(f'done...')
```
